```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from matplotlib import pyplot as plt

X = []
Y = []
```

```python
def f(x):
    # for example let's try to predict values of x^2 
    return x*x

# x values between 0-100 
X = [i for i in range(100)]
# values of y for each x value
Y = [f(i) for i in range(100)]
```

```python
# we define the keras model
model = Sequential([
    Dense(units=1000, input_dim=1, activation='relu'),
    Dense(units=1000, activation='relu'),
    Dense(units=1000, activation='relu'),
    Dense(units=1000, activation='relu'),
    Dense(units=1, activation='linear')
])

# compile the keras model
model.compile(optimizer=Adam(learning_rate=0.01), loss='mean_squared_error', metrics=['accuracy'])

# fit the keras model on the dataset (CPU)
model.fit(X, Y, epochs=1500, shuffle=True)
```

```python
# make class predictions with the model
X = []
X = [i for i in range(100)]
predictions = model.predict(X)
```

```python
plt.figure(figsize=(14,6))
# red dots => real values
# blue dots => predicted values
plt.subplot(121), plt.plot(X, Y,'r.',ls = "dashed")
plt.plot(X, predictions,'b.', ls = "dashed")  , plt.title('Predictions Visualized')
plt.show()
```
